- 일반적으로 우리가 알고있는 선형 회귀 방적식은 데이터 행렬 X의 차원이 1인경우 이다.

- 여기서 입실론은 평균이 베타0인 확률변수이다.

# 다항 회귀 분석

- 데이터 행렬 X의 차원이 1일 때 다항 회귀 방정식은 다음과 같다.
- 단순 회귀 방정식은 직선으로만 표기가 가능하지만, 다항 회귀 방정식은 곡선으로 표현이 가능하여 더 유연히 fitting이 가능하다
- 하지만 항의 갯수를 너무 크게 키우면 분산이 커져 overfitting의 문제가 생길 수 있다.(다중 회귀 방정식도 같은 문제가 발생한다)

- quad fit이 x의 제곱항까지 있는 다항 회귀 방정식이고
- cubic fit이 x의 세제곱항까지 있는 다항 회귀 방정식이다.

# Lasso, Ridge, Elastic Regularization

- overffiting을 줄이기 위한 방법은 다양하지만 여기서는 손실함수에 제약조건을 주어서 정규화를 하는 방법을 활용한다.

# 5. MAE, MSE

- MAE는 예측과 실제값의 차이를 L1 norm을 활용하여 나타낸다.
- MAE는 MSE보다 조금 더 직관적인고 정확한 거리를 준다고 할 수 있다.
- 하지만 MAE는 미분이 불가능하여 최적화를 위한 손실함수로 사용하기가 어렵다

- MSE는 예측과 실제값의 차이를 L2 norm을 활용하여 나타내다.
- MSE는 미분이 가능하여 회귀에 대한 손실함수로 활용할 수 있다.
- 또한 MSE는 차이를 제곱함에따라 손실이 커짐에 따라 손실의 차이가 더 커지게 되어 큰 손실에 대해 가중치를 주는 효과를 낸다.

- 아래는 계수 벡터를 x라고 했을 때 f(x)를 최적화하도록 x를 update하는 식이다.

- 위 식에서 알파는 학습률 이며, 그레디언트 벡터의 방향으로 얼마나 더 x를 이동시킬 것인가에 대한 정도를 의미한다.
- 학습률 알파는 최적화해야할 hyperparameter이다.

- 학습률 알파가 너무 크게 되면 손실함수값이 진동하며 학습이 되지 않을 수 있다.
- 학습률 알파가 너무 작으면 학습 속도가 너무 느리거나 Local Minimum에 쉽게 빠질 수 있다.
- 때문에 처음에는 학습률을 크게하여 Local minimum에 쉽게 빠지지 않으며 빠르게 하강하고 점점 학습을 진행하며 학습률을 줄여나가며 더 섬세하게 global minimum을 찾게끔 한다. - 이러한 방법을 적응형 경사하강법이라고도 부른다.
